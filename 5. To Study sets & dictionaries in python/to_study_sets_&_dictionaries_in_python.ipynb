```python
''' Problem statement 2: Three student choose elective subjects.
find the common subjects chosen by all students using set operations.
concept used: set intersection(&) '''

alice = {"math", "science", "history"}
bob = {"science", "computer science", "history"}
charlie = {"math", "computer science", "history", "science"}

common_subjects = alice & bob & charlie
print(common_subjects)
print("Common Subjects:", common_subjects)
```

```python
''' Problem Statement 3: one set contains students in the cricket club, another in the football club.
find the statements who are:
1. in both clubs
2, in only one club
concept used: intersection, symmetric difference '''
cricketclublist = {"Alice", "Bob", "Charlie", "David"}
footballclublist = {"Bob", "Eve", "Frank", "David"}

print("students in both the clubs:" , cricketclublist & footballclublist)
print("students in only one club:", cricketclublist ^ footballclublist)
```

```python
''' Problem Statement 4: Absent Students
Given a set of all students and a set of present students, find the absent students.
Concepts Used: Set Difference (-)'''

all_students = {"Tanmay", "Drishti", "Jayvee", "Jahnvi", "Tannishtha", "Samiksha", "Rucha"}
present_students = {"Tanmay", "Drishti", "Samiksha", "Jahnvi"}
absent_students = all_students - present_students

print("All Students:", all_students)
print("Present Students:", present_students)
print("Absent Students:", absent_students)
```

```python
''' Problem statement 2: A dictionary stores students' marks. write a python codeto search for a student's marks using the name. if the
student is not found, display a suitable messege '''
student_marks = {
    "Tanmay": 85,
    "Drishti": 92,
    "Jayvee": 78,
    "Jahnvi": 95,
    "Tannishtha": 88
}

print("Student Marks:", student_marks)

student_name_to_search = input("Enter the name of the student to search for: ")

if student_name_to_search in student_marks:
    marks = student_marks[student_name_to_search]
    print(f"{student_name_to_search}'s marks: {marks}")
else:
    print(f"Student '{student_name_to_search}' not found in the records.")
```

```python
''' Problem Statement 3: A dictionary stores usernames and passwords. write a python program to validate user login.'''
users = {"user1": "pass1", "user2" : "pass2", "user3":"pass3"}
user_name = input("Enter your username: ")
password = input("Enter your password: ")
if user_name in users and users[user_name] == password:
    print("Login successful!")
else:
    print("Invalid username or password.")
```

```python
''' Problem Statement 4: A dictionary stores student names and marks. write a python program to find the student with the highest marks.'''
marks= {"Tanmay" : 100, "Tannishtha" : 98, "Rucha" : 90, "Samiksha" : 95}
highest = max(marks, key=marks.get)
print("Student with max marks:", highest, "and Student's Marks: " , marks[highest] )
```
