```python
# list slicing
subjects = ["Physics", "Chemistry", "Maths", "English", "Hindi", "Computer Science"]
print(subjects[1:4]) # 4 is exclusive here
print(subjects[-6:-1])# Reverse Indexing
print(subjects[:4]) # To print without the starting index
print(subjects[2:]) # To print without the ending index
```

```python
# Problem Statement 1 (slicing practice)
numbers = [10, 20, 30, 40, 50]
print('printing elements from 1 to 3')
print(numbers[1:4], "\n")

print('First 3 elements')
print(numbers[:3], "\n")

print('elements from index 2 onwards:(reverse indexing)')
print(numbers[2:], "\n")

print('All elements')
print(numbers[:])
```

```python
# Problem statement 2
marks = [78, 85, 90, 66, 32]
print('Maximum marks')
print(max(marks))
print('minimum marks')
print(min(marks))
print('total marks')
print(sum(marks))
print("average marks")
print(sum(marks)/len(marks))

marks.sort()
print("sorted marks:", marks)
```

```python
# Problem Statement (3 Grocery shopping list )
grocery = ["Dal", "Oil", "Soap", "Masala", "Vegetables"]
grocery1 = ["Tooth Paste", "Tooth Brush", "Biscuits", "Detergent"]

print("added items in the list")
grocery.extend(grocery1)
print(grocery, "\n")

print("removed items from the list")
grocery.remove("Masala")
print(grocery, "\n")

print("the final shopping list")
print(grocery, "\n")
```

```python
# Problem statement 4
contact = ["abc", "def", "ghi", "jkl", "mno"]
contact.append("pqr")
print("The Updated List is: ", contact, "\n")
contact.remove("ghi")
print("The Removed List is: ", contact, "\n")
contact.sort()
print("The sorted List is: ", contact, "\n")
```
