```python
#Dataset
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15],
    'Discount':[10,15,12,20,18,8]
}

df= pd.DataFrame(data)
df
```

```python
# min max normalization

df['price min max'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())

print("\nMin-Max Normalization:")
print(df[['Product','Price', 'price min max']])

```

```python
#2- Z score normalization
df['price z score'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()

print("\nZ-Score Normalization:")
print(df[['Product','Units_Sold', 'price z score']])
```

```python
# normalize multiple columns
cols = ['Price', 'Units_Sold', 'Discount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nNormalized dataset")
print(df_norm)
```

```python
# decimal scaling
df['Price_decimal'] = df['Price']/ 100000
print("\nDecimal Scaling:")
print(df[['Product','Price', 'Prce_decimal']])
```

```python
data = {
    'Order ID': [101, 102, 103, 104, 105, 106],
    'Customer_Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    "Payment_Method": ['Credit Card', 'PayPal', 'Credit Card', 'PayPal', 'Credit Card', 'PayPal'],
    'Product-category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'City':['New York', 'Los Angeles', 'Chicago', 'San Francisco', 'Miami', 'Seattle'],
    'Order_value': [1500, 800, 2000, 1667, 12788, 7756]
}

df1=pd.DataFrame(data)
df1
```

```python
# method 1: label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Gender_label'] = le.fit_transform(df1['Customer_Gender'])
print(df1)
```

```python
# method 2: one hot encoding
df_encoded = pd.get_dummies(df1, columns=['Payment_Method', 'Product-category', 'City'])
df_encoded
```

```python
# method 3: dummy encoding
df_dummy = pd.get_dummies(df1, columns=['Payment_Method', 'Product-category', 'City'], drop_first=True)
df_dummy
```
