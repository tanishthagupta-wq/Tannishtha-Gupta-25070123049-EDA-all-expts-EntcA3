```python
# Problem Statement: get input from user for 5 subjects, find average and calculate grade using if,elif and else
subject1 = int(input("enter the marks of sub1:"))
subject2 = int(input("enter marks of sub2: "))
subject3 = int(input("enter marks of sub3:"))
subject4 = int(input("enter marks of sub4:"))
subject5 = int(input("enter marks of sub5:"))

avg = (subject1 + subject2 + subject3 + subject4 + subject5)/5
print("the average marks of the student is:", avg)

if avg>=90:
  print("student has achieved grade A")
elif avg>=80:
  print("student has achieved grade B")
elif avg>=70:
  print("student has achieved grade C")
elif avg>=60:
  print("student has achieved grade D")
elif avg>=50:
  print("student has achieved grade E")
else:
  print("the student has failed")
```
