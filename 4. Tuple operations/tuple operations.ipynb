```python
# python tuples [tuples are immutable]
# access tuple items
mytodolist= ("eat", "sleep", "repeat")
print(mytodolist[1])
```

```python
mytodolist = ("eat", "sleep", "repeat")
mytodolist[1]= "repair the laptop"
print(mytodolist)
# the error is to show that the list can't be altered
```

```python
# append items
mytodolist = ("eat", "sleep", "repeat")
mytodolist.append("repair the laptop")
print(mytodolist)
# to show that items can neither be added in a list
```

```python
# add tuple to a tuple
mytodolist = ("eat", "sleep", "repeat")
y = ("orange", )
mytodolist += y
print(mytodolist)
```

```python
# slicing a tuple
mytodolist = ("eat", "sleep", "repeat")
print(mytodolist[2:])
```

```python
# problem statement - 1 (tuple practice)
my_tuple = (10, 20, 30, 40, 50)
print('printing elements from 1 to 3')
print(my_tuple[1:4], "\n")

print('First 3 elements')
print(my_tuple[:3], "\n")

print('elements from index 2 onwards:(reverse indexing)')
print(my_tuple[2:], "\n")

print('All elements')
print(my_tuple[:])
```
