```python
#Preprocessing on a Dataset-1
import pandas as pd

data1 = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Tanmay", "Tannishtha", "Kiara", "Adya", "Kashish"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "ENTC", "cse", "CSE"],
    "Admission date": ["01-06-2023"," 03-06-2023", "05-06-2023" , "07-04-2023" , "12-04-2023"]
}
df2 = pd.DataFrame(data1)
df2.to_csv("student_data.csv" , index= False)

```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("student_data.csv")
print("Original data:")
print(df)
```

```python
df.replace("-", np.nan , inplace= True)
df
```

```python
#step 2- replace the - with NaN to mark missing values
df.replace("-", np.nan, inplace= True)
df["Age"] = pd.to_numeric(df["Age"], errors= "coerce")
df["Marks"] = pd.to_numeric(df["Marks"], errors= "coerce")
''' When converting data (like strings -> datetime),
if some values are invalid or cannot be converted,
instead of throwing and error, it will:
Replace those invalid values with NaT (Not a Time)'''

df
```

```python
# step 3, check missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# fill missing age with mean
mean_age = df["Age"].mean()
df["Age"].fillna(mean_age, inplace= True)
```

```python
# fill missing marks with median
median_marks = df["Marks"].median()
df["Marks"].fillna(median_marks, inplace= True)
```

```python
df
```

```python
print(df.isnull(  ).sum())
```

```python
#step 5: fix inconsistent text formatting
df["Department"] = df["Department"].str.upper
```

```python
#step 6: convert admission date to standard date time format
df["Admission date"]  = pd.to_datetime(df["Admission date"] , errors = "coerce")
```

```python
# step 7: show cleaned data
print("\nCleaned data:")
print(df)
```

```python
#step 8 : save cleaned dataset
df.to_csv("cleaned_student_data.csv", index= False)
print("\nCleaned csv file saved as cleaned_student_data.csv")
```
