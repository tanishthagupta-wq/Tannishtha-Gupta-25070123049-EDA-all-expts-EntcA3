```python
''' Problem statement 2:to find the factorial of the entered number by the user'''
n = int(input("Enter the number: "))

if n < 0:
    print("The factorial of the number does not exist")
elif n == 0 or n == 1:
    print("Factorial is 1")
else:
    fact = 1
    i = 1
    while i <= n:
        fact = fact * i
        i = i + 1
    print("Factorial is", fact)
```

```python
 # Armstrong number (153 370 371 407 123) [1^3 + 5^3 + 3^3 = 153]   '''1^3 = 1**3'''
num_str = input("enter the number: ")

if len(num_str) == 3:
    num = int(num_str)
    total = 0
    for digit_char in num_str:
        total += int(digit_char) ** 3   # cube each digit

    if total == num:
        print("It is an Armstrong number")
    else:
        print("It is NOT an Armstrong number")
else:
    print("Please enter a 3-digit number")
```
